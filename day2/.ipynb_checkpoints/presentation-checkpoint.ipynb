# Part 3: Buffer overflows (x86)

The stack:
```
+========================+
|                        |
|  Previous Stack-Frame  |
|      (main func)       |
|                        |
+========================+
|      Parameter 1       |
+------------------------+
|     Return Address     |
+------------------------+
|   Base Pointer (main)  |
+------------------------+ <---- ebp foo
|     Local variables    |
+------------------------+
|     Local variables    | 
+========================+ <---- esp foo
```

# Enter and leave (x86):

- Function call
```as
push   DWORD PTR [ebp-0x8]
call   4ed <foo>
```

- Enter
```as
push   ebp
mov    ebp,esp
```

- Leave (with `leave` or)
```as
mov    esp,ebp
pop    ebp
```

- Function return
```as
ret
```

## Stack canaries


# The [GOT (Global Offset Table)](https://ctf101.org/binary-exploitation/what-is-the-got/)

With dynamically loaded libraries, the GOT contains pointers to the library functions:

-    The GOT contains pointers to libraries which move around due to ASLR
-    The GOT is writable


# ROP Gadgets

- Intel:
    ```
    pop eax ; ret
    xor ebx, ebx ; ret
    ```
- ARM:
    ```
    pop {r4, pc}
    str r1, [r0] ; bx lr
    ```
